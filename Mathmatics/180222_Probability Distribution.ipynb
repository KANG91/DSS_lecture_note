```python
import numpy as np
import scipy as sp
import sklearn as sk
import seaborn as sns
import matplotlib as mpl
mpl.use('Agg')
import matplotlib.pylab as plt
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
```

```python
from scipy import stats
```

```python
mu = 0
std = 1
rv = sp.stats.norm(mu, std)
rv
```

```python
xx = np.linspace(-5, 5 ,100)
plt.plot(xx, rv.pdf(xx))
plt.ylabel("p(x)")
plt.title("pdf of normal distribution")
plt.show()
```

```python
np.random.seed(0)
x = rv.rvs(100)
x
```

```python
sns.distplot(x, kde = False, fit=sp.stats.norm)
plt.show()
```

```python
np.random.seed(0)
x = np.random.randn(100)
plt.figure(figsize = (7,7))
sp.stats.probplot(x, plot = plt)
plt.axis("equal")
plt.show()
```

```python
np.random.seed(0)
x = np.random.rand(100)
plt.figure(figsize = (7, 7))
sp.stats.probplot(x, plot = plt)
plt.ylim(-0.5, 1.5)
plt.show()
```

```python
xx = np.linspace(-2, 2, 100)
plt.figure(figsize = (6, 9))
for i, N in enumerate([1, 2, 3]) :
    X = np.random.rand(1000, N)  - 0.5
    S = X.sum(axis = 1)/np.sqrt(N)
    plt.subplot(3, 2, 2*i + 1)
    sns.distplot(S, bins = 10, kde = False, norm_hist = True)
    plt.xlim(-2, 2)
    plt.yticks([])
    plt.subplot(3, 2, 2* i + 2)
    sp.stats.probplot(S, plot  = plt)
    
plt.tight_layout()
plt.show()
```

```python
xx = np.linspace(0, 1, 1000)
plt.subplot(221)
plt.fill(xx, sp.stats.beta(1.0001, 1.0001).pdf(xx)); plt.ylim(0, 6)
plt.title("(A) a=1, b=1")
plt.subplot(222)
plt.fill(xx, sp.stats.beta(2, 4).pdf(xx)); plt.ylim(0, 6)
plt.title("(B) a=2, b=4, mode={0}".format((4-1)/(4+2-2)))
plt.subplot(223)
plt.fill(xx, sp.stats.beta(8, 4).pdf(xx)); plt.ylim(0, 6)
plt.title("(C) a=8, b=4, mode={0}".format((8-1)/(8+4-2)))
plt.subplot(224)
plt.fill(xx, sp.stats.beta(30, 12).pdf(xx)); plt.ylim(0, 6)
plt.title("(D) a=30, b=12, mode={0}".format((30-1)/(30+12-2)))
plt.tight_layout()
plt.show()
```
