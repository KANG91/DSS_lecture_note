문장내의 인라인 수식은 $\alpha^{\beta}$가 됩니다.

$$ \alpha = \begin{bmatrix} a & b \\ c & d \end{bmatrix} $$
y\combi ^{ -1 }{ x }\quad =\quad w 

```python
from sklearn.datasets import load_boston
boston = load_boston()
x = boston.data
y = boston.target
xinv = np.dot(np.linalg.inv(np.dot(x.T, x)), x.T)

new_x = np.dot(xinv, y)

print(new_x)


```
