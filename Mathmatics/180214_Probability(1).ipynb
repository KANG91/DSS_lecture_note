```python
import numpy as np
import scipy as sp
import sklearn as sk

import matplotlib as mpl
mpl.use('Agg')
import matplotlib.pylab as plt
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
```

```python
import scipy.stats 
```

```python
rv = sp.stats.norm(loc = 1, scale = 2)
```

```python
xx = np.linspace(-8, 8, 100)
pdf = rv.pdf(xx)
plt.plot(xx, pdf)
plt.show()
```

```python
theta = 0.6
rv = sp.stats.bernoulli(theta)
```

```python
xx = [0, 1]
plt.bar(xx, rv.pmf(xx))
plt.xlim(-1, 2)
plt.ylim(0, 1)
plt.xticks([0, 1], ["x=0", "x=1"])
plt.xlabel("sample value")
plt.ylabel("P(x)")
plt.title("pmf of Bernoulli")
plt.show()
```

```python
x = rv.rvs(100, random_state = 0)
x
```

```python
import seaborn as sns
```

```python
sns.countplot(x)
plt.title("Result of Bernoulli Probabiliy Variables")
plt.xlabel("sample value")
plt.show()
```

```python
import pandas as pd
```

```python
y = np.bincount(x, minlength = 2) / float(len(x))

df = pd.DataFrame({"theory" : rv.pmf(xx), "simulation" : y})
df.index = [0, 1]
df
```

```python
df2 = df.stack().reset_index()
df2.columns = ["sample value", "variety", "ratio"]
df2
```

```python
sns.barplot( x = "sample value", y = "ratio", hue = "variety", data = df2)
plt.show()
```

```python
N = 10
theta = 0.6
rv = sp.stats.binom(N, theta)
rv
```

```python
xx = np.arange(N + 1)
plt.bar(xx, rv.pmf(xx), align = "center")
plt.ylabel("P(x)")
plt.title("pmf of binom")
plt.show()
```
