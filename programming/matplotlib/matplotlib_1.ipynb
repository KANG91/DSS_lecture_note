##### 스타일
색상
- https://matplotlib.org/examples/color/named_colors.html
- blue : b
- green : g
- red : r
- cyan : c
- magenta : m
- yellow : y
- black : k
- white : w

마커
- . : point marker
- , : pixel marker
- o : circle marker
- v : triangle_down marker
- ^ : triangle_up marker
- < : triangle_left marker
- `>` : triangle_right marker
- 1 : tri_down marker
- 2 : tri_up marker
- 3 : tri_left marker
- 4 : tri_right marker
- s : square marker
- p : pentagon marker
- `*` : star marker
- h : hexagon1 marker
- H : hexagon2 marker
- `+` : plus marker
- x : x marker
- D : diamond marker
- d : thin_diamond marker

라인 스타일
- `-` : solid line style
- `--` : dashed line style
- `-.` : dash-dot line style
- `:` : dotted line style

커스텀 스타일
- c : 선 색상
- lw : 선 굵기
- ls : 선 스타일
- marker : 마커 종류
- ms : 마커 크기
- mec : 마커 선 색깔
- mew : 마커 선 굵기
- mfc : 마커 내부 색깔

```python
data = np.random.randint(10, size=10)
plt.plot(data, 'rd-')
plt.show()
```

```python
data = np.random.randint(10, size=10)
plt.plot(data, c="r", lw=2, ls="-", marker="4", ms="5", mec="b", mew="5", mfc="y")
plt.show()
```

##### limit 설정

```python
# xlim, ylim 설정
data = np.random.randint(10, size=10)
plt.xlim(-1, 10)
plt.ylim(0, 15)
plt.plot(data)
plt.show()
```

##### 멀티 라인 그리기

```python
data1 = np.random.randint(5, size=5)
data2 = np.random.randint(5, size=5)
data3 = np.random.randint(5, size=5)
xticks = range(0,5)
print(data1, data2, data3)

plt.plot(xticks, data1, 'ro-', xticks, data2, 'g^:', xticks, data3, 'b*--')
plt.show()
```

```python
xticks = range(0,5)
print(data1, data2, data3)

plt.plot(xticks, data1)
plt.plot(xticks, data2)
plt.plot(xticks, data3)
plt.show()
```

##### legend (범례)
범례 위치
- best : 0 (matplot에서 알아서 빈 공간에 범례를 넣어줌)
- upper right : 1
- upper left : 2
- lower left : 3
- lower right : 4
- right : 5
- center left : 6
- center right : 7
- lower center : 8
- upper center : 9
- center : 10

```python
data1 = np.random.randint(5, size=5)
data2 = np.random.randint(5, size=5)
plt.plot(data1, label="data4")
plt.plot(data2, label="data2")
plt.legend(loc=0)
plt.show()
```

```python
data1 = np.random.randint(5, size=5)
data2 = np.random.randint(5, size=5)
plt.plot(data1, label="data1")
plt.plot(data2, label="data2")
plt.legend(loc=1)
plt.show()
```

### Figure(그래프 바깥 쪽의 영역)
- 그래프의 크기를 조절

##### 그래프 크기 조절

```python
data = np.random.randint(5, size=5)
f1 = plt.figure(figsize=(10, 5)) #(가로, 세로)
plt.plot(data)
plt.show()
```

##### 여러개 그래프 출력

```python
data1 = np.random.randint(5, size=5)
data2 = np.random.randint(5, size=5)
print(data1, data2)

plt.subplot(3, 1, 1) # 2*1에서 1번 위치
plt.plot(data1)

plt.subplot(3, 1, 3) # 2*1에서 2번 위치
plt.plot(data2)

plt.show()
```

```python
data1 = np.random.randint(5, size=5)
data2 = np.random.randint(5, size=5)
data3 = np.random.randint(5, size=5)
data4 = np.random.randint(5, size=5)

print(data1, data2, data3, data4)

plt.subplot(221) # 2*2에서 1번 위치
plt.plot(data1)

plt.subplot(222) # 2*2에서 2번 위치
plt.plot(data2)

plt.subplot(223) # 2*2에서 3번 위치
plt.plot(data3)

plt.subplot(224) # 2*2에서 4번 위치
plt.plot(data4)

plt.show()
```
