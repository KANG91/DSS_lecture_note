```python
import pandas as pd
import numpy as np
number = 10
names = ['Adam', 'Alan', 'Alex', 'Alvin', 'Andrew', 'Anthony', 'Arnold', 'Baldy', 
         'Baron', 'Billy', 'Boris', 'Bruno', 'Caley', 'Champ', 'Charlie', 'Clark']

df = pd.DataFrame(columns=["Name","Math","Science","English"])

# set data
def get_name() :
    return np.random.choice(names)

def get_score() :
    return np.random.randint(50, 100 + 1)

def get_data(number) :
    df = pd.DataFrame(columns = ["Name", "English", "Science", "Math"])
    for _ in range(number) :
        name = get_name()
        
        while name in list(df["Name"]) :
            name = get_name()
            
        dic = {
            "Name" : name,
            "English" : get_score(),
            "Math" : get_score(),
            "Science" : get_score(),
        }
       
        df.loc[len(df)] = dic
    return df

def get_grade(point):
    if point >= 270:
        return "A"
    elif point > 240:
        return "B"
    elif point >= 210:
        return "C"
    elif point >= 180:
        return "D"
    else:
        return "F"

# set total
df["Total"] = df["English"] + df["Science"] + df["Math"]

# set grades
```

```python
df = get_data(10)
df["Total"] = df["English"] + df["Science"] + df["Math"]
df
df['Grade'] = df['Total'].apply(get_grade)
df.sort_values(by = ["Grade"], ascending = True)
```
